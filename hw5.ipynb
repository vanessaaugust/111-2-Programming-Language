```python
import pandas as pd
from collections import Counter
data = pd.read_csv('Singapore_Listings.csv')
data.head()
```

```python
subdata = data[['name','neighbourhood_group', 'neighbourhood']]
subdata
```

```python
a = subdata.groupby(['name'])['neighbourhood'].apply(lambda x: ','.join(x)).reset_index()
a
```

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = a['neighbourhood'].tolist()
vec = CountVectorizer()
X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df
```

```python
coocc = df.T.dot(df)
coocc
```

```python
import numpy as np

filter_num = 20
getid = coocc.index
important = pd.DataFrame(columns = getid).T
important['relation'] = ""
subcoocc = coocc
dropid = []

for i in range(len(coocc)):
    result = coocc[coocc.iloc[i] > filter_num].index.tolist()    
    if(len(result) > 0):
        important['relation'].loc[getid[i]] = coocc[coocc.iloc[i] > filter_num].index.tolist()
    else:
        dropid.append(i)
        important = important.drop([getid[i]])
        subcoocc = subcoocc.drop([getid[i]])    
subcoocc
```

```python
subcoocc.columns[dropid]
```

```python
subcoocc.drop(subcoocc.columns[dropid], axis=1, inplace=True)
subcoocc
```
